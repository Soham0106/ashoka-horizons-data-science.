```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
pima=pd.read_csv("/Users/atulagrawal/Downloads/pima-2/diabetes.csv")
starbucks=pd.read_csv("/Users/atulagrawal/Downloads/starbucks/starbucks.csv")
netflix=pd.read_csv("/Users/atulagrawal/Downloads/netflix/netflix_titles.csv")
```

```python
fig,ax=plt.subplots()
ax.hist(starbucks["Calories"].values, bins=20, color="skyblue", edgecolor="black")
ax.set_title("Calories Distribution")
ax.set_xlabel("Calories")
ax.set_ylabel("Frequency")
plt.show()
```

```python
fig,ax=plt.subplots()
sns.boxplot(starbucks["Calories"].values,ax=ax)
ax.set_title("Box Plot for Calories")
ax.set_ylabel("Mean Value")
ax.set_xticks([])
plt.show()
```

```python
#The average calories is estimated around 200. For items in the middle of list, the ampunt if calories are large, concentrating around 125 and 270.
pima
```

```python
fig,ax=plt.subplots()
sns.countplot(data=netflix, y="type", palette=["#511D43","#9BC09C"], ax=ax)
ax.set_title("CountPlot for Type")
ax.set_xlabel("Frequency")
ax.set_ylabel("Type")
plt.show()
```

```python
#Movies is the most common type
```

```python
fig,ax=plt.subplots()
sns.histplot(data=pima, x="Age", color="cyan", ax=ax, kde=True, bins=20)
ax.set_title("Age Distribution by pima")
ax.set_xlabel("Age Bins")
ax.set_ylabel("Frequency")
plt.show()
```

```python
fig, ax=plt.subplots()
sns.scatterplot(data=starbucks, x="Calories", y=" Sugars (g)", ax=ax, color="#0B1D51")
ax.set_xlabel("Calories")
ax.set_ylabel("Sugar(g)")
ax.set_title("Relationship between Calories and Sugar")
plt.show()
```

```python
#Positive Correlation. More the sugar, more the calories consumed
```

```python
fig, ax=plt.subplots()
sns.violinplot(data=pima, x="Outcome", y="Glucose", ax=ax)
ax.set_title("Glucose to Outcome")
ax.set_ylabel("Glucose")
ax.set_xlabel("Outcome")
plt.show()
```

```python
#The people with higher agregate glucose are more prone to diabetes and have generally higher concentrations of glucose
```

```python
fig, ax=plt.subplots()
a=netflix.groupby("release_year")["title"].count()
sns.lineplot(data=a, ax=ax)
plt.title('Total Number of Netflix Titles Released Each Year')
plt.xlabel('Release Year')
plt.ylabel('Number of Titles')
plt.show()
```

```python
fig,ax=plt.subplots()
release_counts = netflix.groupby(['release_year', 'type']).size().reset_index(name='count')
sns.lineplot(data=release_counts, x="release_year", y="count", hue="type", ax=ax)
plt.title('Number of Releases by Year')
plt.xlabel('Release Year')
plt.ylabel('Number of Titles')
plt.legend(title='Type')
plt.show()
```

```python
fig, ax=plt.subplots()
sns.scatterplot(data=pima, x="Age", y="BMI", hue="Outcome", size="Pregnancies")
plt.title("BMI vs Age")
plt.xlabel("Age")
plt.ylabel("BMI")
plt.show
```

```python
#Higher BMI values and age and BMI, Higher chances of being diabetic
#Correlation between Pregnancies and Diabetes risk
```

```python
feat=["Pregnancies", "Glucose", "BloodPressure", "BMI", "Age"]
mat=pima[feat].corr()
plt.figure()
sns.heatmap(mat, annot=True)
plt.title("Correlations")
plt.show()
```

```python
#High correlation-Age with Pregnancies
#Low correlation-BMI with Pregnancies
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2)


sns.histplot(data=pima, x="BloodPressure", kde=True, ax=ax[0, 0])
ax[0, 0].set_title("Distribution of Blood Pressure")
ax[0, 0].set_xlabel("BP")
ax[0, 0].set_ylabel("Frequency")

sns.scatterplot(data=starbucks, x="Calories", y=" Protein (g) ", ax=ax[0, 1])
ax[0, 1].set_title("Calories vs Protein")
ax[0, 1].set_xlabel("Calories")
ax[0, 1].set_ylabel("Protein(g)")

sns.countplot(data=starbucks, x="Beverage_category", ax=ax[1,0])
ax[1, 0].set_title("Beverage Count")
ax[1, 0].set_xlabel("Company")
ax[1, 0].set_ylabel("Count")
ax[1, 0].tick_params(axis='x', rotation=90) 

sns.boxplot(data=pima, x="Outcome", y="Age", ax=ax[1,1])
ax[1, 1].set_title("Age by Outcome")
ax[1, 1].set_xlabel("Result")
ax[1, 1].set_ylabel("Age")
```
